## Training the Data

## 1.Using Linear Regression

```python
X=data.drop("income",axis=1)
y=data['income']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
predict = logmodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predict)*100
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predict))
```

## 2.Using Stochastic Gradient Descent

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
a = cross_val_score(sgd_clf, X_train, y_train, cv=5, scoring="accuracy")
a
```

```python
a.mean()*100
```

```python
predict_sgd = sgd_clf.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predict_sgd))
```

## 3.Using Gaussian Naive Bayes

```python
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
b= cross_val_score(model, X_train, y_train, cv=5, scoring="accuracy")
b
```

```python
b.mean()
```

```python
predict = model.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predict))
```

## 4. Using Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(max_leaf_nodes=10, random_state=0)
dtc.fit(X_train, y_train)
predicted = dtc.predict(X_test)
accuracy_score(y_test,predicted)*100
```

```python
print(classification_report(y_test,predicted))
```

## 5.Using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=30)
rfc.fit(X_train,y_train)
predicting = rfc.predict(X_test)
```

```python
print(classification_report(y_test,predicting))
```
